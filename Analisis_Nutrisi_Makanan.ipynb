```python
df = df.astype({'Calories': int, 'Carbs': int, 'Fats': int, 'Protein': int})
```

**EDA**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi menggunakan histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['Calories'], bins=20, kde=True)
plt.title('Distribusi Kalori')
plt.xlabel('Calories')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Menghitung matriks korelasi hanya untuk kolom numerik
correlation_matrix = df.select_dtypes(include=np.number).corr()

# Menampilkan matriks korelasi menggunakan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi')
plt.show()
```

```python
plt.figure(figsize = (10,4))
plt.title("Category Name",fontsize = 8)
ax = sns.countplot(x = "Category Name",data = df,hue = "Category Name")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=8)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Fats', y='Calories', data=df)
plt.title('Korelasi antara Kalori dan Lemak')
plt.xlabel('Lemak')
plt.ylabel('Kalori')
plt.show()
```

```python
top_calories = df.sort_values(by=['Calories'], ascending=False).head()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Category Name', y='Calories', data=top_calories)
plt.title('Kategori Makanan Tinggi Kalori')
plt.xlabel('Kategori')
plt.ylabel('Kalori')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()

# Menambahkan anotasi nilai kalori di atas setiap bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),  # Format nilai kalori
                (p.get_x() + p.get_width() / 2., p.get_height()),  # Koordinat anotasi
                ha='center', va='center',  # Perataan anotasi
                xytext=(0, 9),  # Offset anotasi
                textcoords='offset points')  # Sistem koordinat offset

plt.show()
```

```python
# Urutkan data berdasarkan kalori tertinggi dan ambil 10 teratas
top_calories_foods = df.sort_values(by=['Calories'], ascending=False).head(10)

# Buat bar plot
plt.figure(figsize=(12, 6))  # Atur ukuran gambar
sns.barplot(x='Food Name', y='Calories', data=top_calories_foods, palette='viridis')
plt.title('Top 10 Makanan dengan Kalori Tertinggi', fontsize=16)  # Judul grafik
plt.xlabel('Nama Makanan', fontsize=12)  # Label sumbu x
plt.ylabel('Kalori', fontsize=12)  # Label sumbu y
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotasi label sumbu x agar mudah dibaca
plt.tight_layout()  # Sesuaikan tata letak agar label tidak terpotong
plt.show()  # Tampilkan grafik
```
